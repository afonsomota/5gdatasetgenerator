```python
import glob
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import scipy
import scipy.stats
```

```python
times = {}
packets = {}
for f_name in glob.glob("*.csv"):
    times[f_name] = []
    packets[f_name] = []
    for line in open(f_name).readlines():
        t = float(line.split(",")[0])
        l = float(line.split(",")[1])
        times[f_name].append(t)
        packets[f_name].append(l)

for k in times:
    times[k] = np.array(times[k])
    packets[k] = np.array(packets[k])
    
```

```python
compiled = pd.Series(dtype="float32")
out = None
for f_name in times:
    #t_pd = pd.Series(times[f_name]).diff()[1:]
    t_pd = pd.Series(times[f_name])
    p_pd = pd.Series(packets[f_name])
    out = t_pd
    compiled = pd.concat([compiled, t_pd])
    plt.title(f_name)
    #sns.distplot(t_pd, hist = False, kde = True,
    #             kde_kws = {'linewidth': 1},
    #             label = f_name)
    dist = scipy.stats.lognorm
    param = dist.fit(t_pd)
    size=max(t_pd)
    x=np.linspace(0,size,500)
    plt.plot(t_pd, p_pd, label=f_name)
    #pdf_fitted = dist.pdf(x, *param[:-2], loc=param[-2], scale=param[-1]) 
    #plt.plot(x,pdf_fitted,label="fitted")
    plt.show()


dist = scipy.stats.lognorm
param = dist.fit(compiled)
size=max(compiled)
x=np.linspace(0,size,500)
pdf_fitted = dist.pdf(x, *param[:-2], loc=param[-2], scale=param[-1]) 

sns.distplot(compiled, hist = False, kde = True,
                 kde_kws = {'linewidth': 1},
                 label = "compiled")
#plt.plot(x,pdf_fitted,label="fitted")
```
